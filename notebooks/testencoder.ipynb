```python
from torch.utils.data import Dataset
import torchvision

dataset = torchvision.datasets.Caltech101(root='../data', download=True)
```

# Encoder with Projection into (1, 512)

```python
from torch import nn
import lightning as L
from transformers import AutoProcessor, CLIPVisionModelWithProjection

# Image processor for encoder
processor = AutoProcessor.from_pretrained('openai/clip-vit-base-patch32')

# Image encoder used by CLIP
encoder = CLIPVisionModelWithProjection.from_pretrained('openai/clip-vit-base-patch32')
# Freeze model
for param in encoder.parameters():
    param.requires_grad = False
```

```python
image, label = dataset[0]
inputs = processor(images=image, return_tensors='pt')
outputs = encoder(**inputs)
print(outputs)
```
